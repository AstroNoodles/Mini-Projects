### Imputing

```python
import random
import numpy as np

items = [random.randint(1, 100) if i % 2 == 0 else np.nan for i in range(150)]
print(len(items))

df['extra'] = items
items
```

```python
df.head()
```

```python
from sklearn.impute import SimpleImputer
```

```python
impute = SimpleImputer(missing_values=np.nan, strategy="median")
impute.fit(df[['extra']])

df['extra'] = impute.transform(df[['extra']])
df.head()
```

```python
del df['extra']
```

```python
df.head()
```

## Simple ML

```python
# Classification Time with a SVM
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df.drop('target', axis=1), df['target'])
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

print(f'--> Accuracy Score: {accuracy_score(y_test, y_pred)}')
print()
print(f'Precision Score: {precision_score(y_test, y_pred, average ="macro")}')
print(f'Recall Score: {recall_score(y_test, y_pred, average ="macro")}')
print(f'--> F1 Score: {f1_score(y_test, y_pred, average ="macro")}')
```
